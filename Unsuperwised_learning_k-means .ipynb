# no train_split, direct test data

```python
x = ds.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
```

```python
no_of_cluster =16
wcss_array = []


for i in range(1,no_of_cluster):
    k_means = KMeans(n_clusters = i, init = 'k-means++',random_state =42)
    k_means.fit(x)
        #print(k_means.inertia_)   # cheack wcss(within cluster sum of square) values
    wcss_array.append(k_means.inertia_)
print(wcss_array)  # we are draw the graph so we find wcss_array values in array

    
```

```python
# example
no_of_cluster2 =200
wcss_array2 = []

for i in range(1,no_of_cluster2):
    k_means2 = KMeans(n_clusters = i,init ='k-means++',random_state = 42)
    k_means2.fit(x)
    #print(k_means2.inertia_)
    wcss_array2.append(k_means2.inertia_)
print(wcss_array2)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,no_of_cluster),  wcss_array)
plt.title('Elboe Method')
plt.xlabel('n of clusters')
plt.ylabel('wcss score or result')
plt.show()
```

# so we find best no. of cluster from elbow method k = 5

```python
# so we apply the use of cluster value 5
k_means = KMeans(n_clusters = 5,init = 'k-means++',random_state =50)
kmeans_pred = k_means.fit_predict(x)
kmeans_pred
```
