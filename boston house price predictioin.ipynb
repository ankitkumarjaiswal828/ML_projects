```python
df = pd.read_csv('C:\\Users\\Ankit\\Desktop\\564265_1025037_bundle_archive\\boston.csv')
df
```

```python
print(df.isnull().values.any())
```

```python
import matplotlib.pyplot as plt
def function(df,size=12):
    corr =df.corr()
    fig , ax =plt.subplots(figsize=(size,size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)),corr.columns)
    plt.yticks(range(len(corr.columns)),corr.columns)
function(df)
```

```python
#saving the data into x and y form
x = df.iloc[:, :-1].values
y = df.iloc[:,-1].values
y # y is dependent variables results (prices)
x# x is independent variables upto (unnamed  to  lstat)
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test , y_train ,y_test = train_test_split(x ,y , test_size =0.33  ,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
print(lin_reg.score(x_train,y_train))
```

```python
print(lin_reg.score(x_test, y_test))
```

```python
y_predict = lin_reg.predict(x_test)
y_predict
```

```python
#Check R-square value on training data
print("R-Square Training Set: {0}".format(lin_reg.score(x_train, y_train)))
```

```python
#Check R-square value on test data
print("R-Square Training Set:{0}".format(lin_reg.score(x_test  ,y_test)))
```
